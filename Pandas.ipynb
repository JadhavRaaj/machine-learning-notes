# **PANDAS**

Pandas is the powerful data manipulation library widely used for data analysis and data cleaning. it provides two primary data structures i.e. **series and data frame**. a **series is one dimentional array like object.** while **data frame is two dimentional size-mutable** and potentially heterogenous tabular data structure with labelled as axes(rows and columns).

```python
#SERIES =
#PANDAS SERIES IS SIMILAR TO ONE DIMENSIONAL ARRAY LIKE OBJECT. THAT CAN STORE ANY DATA TYPE. IT IS SIMILAR TO COLUMN IN THE TABLE.
```

```python
import pandas as pd
data = [1,2,3,4,5]
series = pd.Series(data)
print(series)
```

```python
#create series from dictionary
import pandas as pd
data = {'a':1,'b':2,'c':3,'d':4,'e':5}
series_dic = pd.Series(data)
print(series_dic)
```

```python
data = [1,2,3,4,5]
index = ['a','b','c','d','e']
series_index = pd.Series(data,index=index)
print(series_index)
```

```python
#DATFRAME
#create a dataframe from a dictionary of list
data = {
    'name':['alice','bob','charlie'],
    'age':[25,30,35],
    'city':['new york','san francisco','los angeles']
}
df = pd.DataFrame(data)
print(df)
print(type(df))
```

```python
#DATA MANIPULATION WITH DATA FRAME
```

```python
#ADDING A COLUMN
df['salary in k']=[200,500,1000]
df
```

```python
#remove a column temporarily
df.drop('salary in k',axis=1)# axis=1 beacuse we have to drop a column by default axis=0 ie. it checks for row and if we not explicitly involve axis we will get error
#to make this drop permanant we have to include 'inplace=true' to save the state.
```

```python
df.drop('salary in k', axis=1, inplace=True)
df
```

```python
df['age']=df['age']+1
df
```

```python
#dropping a row permanantly
df.drop(2,axis=0,inplace=True)
df
```

```python
df.drop(2,axis=1,inplace=True)
df
```
