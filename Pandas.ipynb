# **PANDAS**

Pandas is the powerful data manipulation library widely used for data analysis and data cleaning. it provides two primary data structures i.e. **series and data frame**. a **series is one dimentional array like object.** while **data frame is two dimentional size-mutable** and potentially heterogenous tabular data structure with labelled as axes(rows and columns).

```python
#SERIES =
#PANDAS SERIES IS SIMILAR TO ONE DIMENSIONAL ARRAY LIKE OBJECT. THAT CAN STORE ANY DATA TYPE. IT IS SIMILAR TO COLUMN IN THE TABLE.
```

```python
import pandas as pd
data = [1,2,3,4,5]
series = pd.Series(data)
print(series)
```

```python
#create series from dictionary
import pandas as pd
data = {'a':1,'b':2,'c':3,'d':4,'e':5}
series_dic = pd.Series(data)
print(series_dic)
```

```python
data = [1,2,3,4,5]
index = ['a','b','c','d','e']
series_index = pd.Series(data,index=index)
print(series_index)
```

```python
#DATFRAME
#create a dataframe from a dictionary of list
data = {
    'name':['alice','bob','charlie'],
    'age':[25,30,35],
    'city':['new york','san francisco','los angeles']
}
df = pd.DataFrame(data)
print(df)
print(type(df))
```

```python
#merging and joining dataframe
df1 = pd.DataFrame({'key':['a','b','c','e'],'value':[1,2,3,4]})
df2 = pd.DataFrame({'key':['a','b','c','d'],'value':[5,6,7,8]})
#merging
merged_df = pd.merge(df1,df2,on='key',how = "outer")#this 'how' field decides the type of join you want to perform inner,outer etc
merged_df
#lets try performing the above operations
merged_df['value_x_filled']= merged_df['value_x'].fillna(merged_df['value_x'].mean())
merged_df['value_y_filled']= merged_df['value_y'].fillna(merged_df['value_y'].mean())
merged_df
#change the data type of some column
merged_df['value_x_filled_int']= merged_df['value_x_filled'].astype(int)
merged_df['value_y_filled_int']= merged_df['value_y_filled'].astype(int)
merged_df
#change the column name
merged_df = merged_df.rename(columns={'value_x_filled':'value_x_filled_new'})
merged_df = merged_df.rename(columns={'value_y_filled':'value_y_filled_new'})
merged_df
#perform aggregate function
grouped_agg = merged_df.groupby('key')['value_x_filled_new'].agg(['mean','max','min'])
grouped_agg
merged_df.describe()
```
