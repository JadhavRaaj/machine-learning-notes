# **Focal Loss**

**The Problem:** The Lazy AI
Imagine you are building an AI to detect a rare disease. You have pictures of 99,000 healthy patients and only 1,000 sick patients. This is what we call an imbalanced dataset.

If you use the standard way of training an AI (which is called Cross-Entropy), the AI quickly figures out a cheat code: "If I just guess 'Healthy' every single time, I will be 99% accurate!"

The AI gets lazy. It stops trying to learn what the disease actually looks like. Why? Because the tiny amount of blame it gets for missing the 1,000 sick patients is completely covered up by the massive amount of praise it gets for correctly guessing the 99,000 healthy patients.

**The Solution:** Focal Loss (The Strict Teacher)
Focal Loss fixes this lazy behavior by acting like a very strict teacher. It changes how the AI is scored.

When the AI gets an easy example right (like a clearly healthy patient), Focal Loss says: "That was obvious. I am shrinking your score for this to almost zero. Stop focusing on this."

When the AI gets a hard or rare example wrong (like missing a sick patient), Focal Loss says: "Pay attention! This one matters! I am keeping the penalty high so you learn from this mistake."

**The Advantage:** Instead of being overwhelmed by thousands of easy, boring examples, the AI is forced to focus all of its brainpower on the hard, rare examples. It stops guessing and actually learns the difficult stuff.

```python
import numpy as np
```

```python
def standard_loss(confidence):
    """
    Standard Cross-Entropy Loss.
    It penalizes the AI based on how confident it is.
    """
    # Prevent math errors with a tiny number
    confidence = np.clip(confidence, 1e-7, 1. - 1e-7)
    return -np.log(confidence)
```

```python
def focal_loss(confidence, strictness=2.0):
    """
    Focal Loss (The Strict Teacher).
    It looks at the standard loss and shrinks it if the AI was already confident.
    """
    confidence = np.clip(confidence, 1e-7, 1. - 1e-7)
    standard_penalty = -np.log(confidence)

    # THE MAGIC HAPPENS HERE:
    # If confidence is high (e.g., 0.9), (1 - 0.9) is 0.1.
    # 0.1 squared is 0.01. The penalty is drastically shrunk!
    shrink_factor = (1 - confidence) ** strictness

    return shrink_factor * standard_penalty
```

```python
# --- Let's run a test to see the difference ---

print("SCENARIO 1: The Easy Example")
print("The AI looks at a healthy patient and is 90% sure (0.90) they are healthy.")
easy_confidence = 0.90

print(f"Standard AI Penalty: {standard_loss(easy_confidence):.4f}")
print(f"Focal Loss Penalty:  {focal_loss(easy_confidence):.4f}")
print("-> Notice how Focal Loss shrinks the penalty to almost zero! It ignores easy stuff.\n")


print("SCENARIO 2: The Hard, Rare Example")
print("The AI looks at a sick patient but is only 10% sure (0.10) they are sick. It's struggling.")
hard_confidence = 0.10

print(f"Standard AI Penalty: {standard_loss(hard_confidence):.4f}")
print(f"Focal Loss Penalty:  {focal_loss(hard_confidence):.4f}")
print("-> Notice how Focal Loss keeps this penalty very high. It forces the AI to learn!\n")
```
